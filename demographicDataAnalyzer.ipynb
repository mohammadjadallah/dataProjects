### What is the percentage of people who have a Bachelor's degree?
<br>

To find the percentage we must (devide each Xi on sum of observations) * 100

(column / column.sum()) * 100

df[percent] = (df['column_name'] / df['column_name'].sum()) * 100
<br>

```python
df['education']  # education column data
```

```python
df.education[df['education'] == "Bachelors"]
```

```python
# solution of question
# or this way >> round((df.loc[df['education'] == "Bachelors"].size / df.size) * 100, 1)
B_percentage = round((df.education[df['education'] == "Bachelors"].size / df.education.count()) * 100, 1)
B_percentage
```

### What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

```python
BMD = df[(df['education'] == "Bachelors") | (df['education'] == "Masters") |
                                     (df['education'] == "Doctorate")]

BMD_percentage = round((BMD['education'][df['salary'] == ">50K"].size / BMD['education'].size) * 100, 1)
BMD_percentage
```

### What percentage of people without advanced education make more than 50K?
<br>

```python
df.education.value_counts()
```

```python
not_higher1  = df[(df['education'] != "Bachelors")]
not_higher2 = not_higher1[(df['education'] != "Masters")]
not_higher3 =  not_higher2[(df['education'] != "Doctorate")]
not_higher = not_higher3.education[not_higher3.salary == ">50K"]
# not_higher.education.value_counts()
not_higher_percentage = round((not_higher.size / not_higher3.education.size) * 100, 1)
not_higher_percentage
```

```python
# isin() هذه الطريقة أفضل
# df.education[df.education.isin(['Masters'])]
# ~ >> not in
nh = df[~df.education.isin(['Masters', 'Bachelors', 'Doctorate'])]
new_nh = nh[nh.salary == '>50K']

nh_percentage = round((new_nh.size / nh.size) * 100, 1)

nh_percentage
```

### What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?
<br>

```python
with_salary_more_50k = df['hours-per-week'][df['hours-per-week'] == min_hours][df['salary'] == ">50K"]
with_all_salary = df['hours-per-week'][df['hours-per-week'] == min_hours]

percentage_work_minimum_hours_per_week = round((with_salary_more_50k.size / with_all_salary.size) * 100, 1)
percentage_work_minimum_hours_per_week
```

### What country has the highest percentage of people that earn >50K and what is that percentage?
<br>

```python
df
```

```python
higher_salary = df.groupby(df['native-country']).apply(
    lambda data: (data[data['salary'] == ">50K"].size / data.size) * 100
    )

higher_salary_country = higher_salary.idxmax()
higher_salary_country

hiegher_salary_percentage = round(higher_salary.max(), 1)
hiegher_salary_percentage
```

```python
# all

def demographic_analayzer(print_=True):
    # read the csv file
    df = pd.read_csv("adult.data.csv")
    race_counts = df['race'].groupby(df['race']).count()  # or this way >> df['race'].value_counts()
    
    # give me age for each Male then calculate the average of them "Men"
    avg = round(df['age'][df['sex'] == "Male"].mean(), 1)
    
    B_percentage = round((df.education[df['education'] == "Bachelors"].size / df.education.count()) * 100, 1)
    
    BMD_percentage = round((BMD['education'][df['salary'] == ">50K"].size / BMD['education'].size) * 100, 1)

    # ~ >> not in
    nh = df[~df.education.isin(['Masters', 'Bachelors', 'Doctorate'])]
    new_nh = nh[nh.salary == '>50K']

    nh_percentage = round((new_nh.size / nh.size) * 100, 1)

    min_hours = df['hours-per-week'].min()
    
    with_salary_more_50k = df['hours-per-week'][df['hours-per-week'] == min_hours][df['salary'] == ">50K"]
    with_all_salary = df['hours-per-week'][df['hours-per-week'] == min_hours]

    percentage_work_minimum_hours_per_week = round((with_salary_more_50k.size / with_all_salary.size) * 100, 1)
    
    higher_salary = df.groupby(df['native-country']).apply(
    lambda data: (data[data['salary'] == ">50K"].size / data.size) * 100
    )
    higher_salary_country = higher_salary.idxmax()
    hiegher_salary_percentage = round(higher_salary.max(), 1)
    
    most_occupation = df.occupation[(df["salary"] == ">50K") & (df["native-country"] == "India")]
    most_occupation = most_occupation.value_counts().idxmax()  # idxmax() will give us ths string not the number like max()
    
    
    if print_:
        
        print("Number of each race:\n", race_counts)
        print("Average age of men:", avg)
        print(f"Percentage with Bachelors degrees: {B_percentage}%")
        print(
            f"Percentage with higher education that earn >50K: {BMD_percentage}%"
        )
        print(
            f"Percentage without higher education that earn >50K: {nh_percentage}%"
        )
        print(f"Min work time: {min_hours} hours/week")
        print(
            f"Percentage of rich among those who work fewest hours: {percentage_work_minimum_hours_per_week}%"
        )
        print(
            "Country with highest percentage of rich:", higher_salary_country
        )
        print(
            f"Highest percentage of rich people in country: {hiegher_salary_percentage}%"
        )
        print("Top occupations in India:", most_occupation)

        
    return {
        "race_count": race_counts,
        "average_age_men": avg,
        "percentage_bachelors": B_percentage,
        "higher_education_rich": BMD_percentage,
        "lower_education_rich": nh_percentage,
        "min_work_hours": min_hours,
        "rich_percentage": percentage_work_minimum_hours_per_week,
        "highest_earning_country": higher_salary_country,
        "highest_earning_country_percentage": hiegher_salary_percentage,
        "top_IN_occupation": most_occupation,
    }
        
        
```
